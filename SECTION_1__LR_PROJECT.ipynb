```python
# Tutorial 1
import pandas as pd
```

```python
# Read a .csv file into a pandas dataframe from a URL

url = ""
dataset = pd.read_csv(url)
```

```python
#See the first 5 rows of your pandas dataframe

dataset.head()
```

```python
# Tutorial 2
#Print the dimensions of your pandas dataframe

dataset.shape
```

```python
#Print information about your pandas dataframe

dataset.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Draw a scatter plot between two variables from your dataset

plt.figure(figsize=(8, 4))

sns.scatterplot(data=dataset, x="Fuel-Consumption-City", y="CO2-Emissions").set(title = "Correlation between CO2-Emissions and Fuel-Consumption-City")
```

```python
#Check pairwise relationships between variables

sns.pairplot(dataset,x_vars=["Engine-Size","Cylinders", "Fuel-Consumption-Hwy", "Fuel-Consumption-Comb"],y_vars = "CO2-Emissions", dropna = True)
```
