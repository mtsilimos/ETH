```python
#the students could get a sense of the basic concept/Assumption 1 of Simple Linear Regression, that is, correlation between an independent and dependent variable
#missing values would not worry them at this stage

sns.pairplot(dataset,x_vars=["LotArea","GrLivArea", "MSSubClass", "1stFlrSF", "2ndFlrSF", "GarageYrBlt"],y_vars = "SalePrice", height=3, dropna = True)
```

```python
#we could also have a categorical variable  to see trends and distributions for its categories
#The pairplot is coloured by each of the categories within that categorical variable (MSZoning)
sns.pairplot(dataset,x_vars=["LotArea","GrLivArea", "MSSubClass", "1stFlrSF", "2ndFlrSF", "GarageYrBlt"],y_vars = "SalePrice", hue = "MSZoning", height=3, dropna = True)
```
